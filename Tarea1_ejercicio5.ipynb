El sistema de gestión de becas
Para cerrar la serie, construirás un sistema interactivo básico que recopile y procese
información escolar utilizando un ciclo while y un menú en consola.
1. Define un diccionario inicial llamado alumnos donde las llaves sean los
nombres de tus estudiantes y los valores sean sus promedios generales (ej.
{"Ana": 9.2, "Luis": 7.8, "Carlos": 8.5}).
2. Diseña un menú interactivo dentro de un ciclo while con las siguientes
opciones para el usuario:
o Opción 1 (Agregar/Modificar): Solicita el nombre de un alumno y su
promedio. Agrega el registro al diccionario o actualízalo si ya existía.
o Opción 2 (Candidatos a Beca): Utiliza una list comprehension para
revisar el diccionario de alumnos y generar una lista que contenga
únicamente los nombres de los estudiantes con un promedio mayor o
igual a 8.5. Muestra la lista en pantalla.
o Opción 3 (Bono de Apoyo): El consejo escolar aprueba un apoyo de 0.3
puntos a todos los alumnos. Utiliza una función lambda (puedes
apoyarte en un ciclo o en map()) para incrementar en 0.3 el promedio
de cada alumno en el diccionario (asegúrate de que ningún promedio
supere la calificación máxima de 10.0).
o Opción 4 (Salir): Termina la ejecución del ciclo while con un mensaje
de despedida.

```python
#definimos diccionario
alumnos = {
    "Carlos" : 9,
    "Maria" : 8.5,
    "José" : 7.8,
    "Pedro" : 9.8,
    "Julio" : 7,
    "Carolina" : 8.5,
    "Pablo" : 7.6
}

pregunta = input("Elige una opción entre: agregar/modificar, candidatos a beca, bono de apoyo")

while pregunta != "salir":

    #primera opción del menú
    if pregunta == "agregar" or pregunta == "modificar":
        #preguntamos el nombre
        nombre = input("Escribe tu nombre: ")
        #preguntamos el promedio
        promedio = int(input("Escribe tu promedio: "))
        alumnos[nombre] = promedio #lo crea si no existe y lo reeemplaza si ya existía
        print(f"ahora tu promedio se ha modificado por un {promedio}") #esto lo puse para saber que sí se había modificado
        #volvemos a preguntar para que hay un bucle
        pregunta = input("Elige una opción entre: agregar/modificar, candidatos a beca, bono de apoyo")

    # segunda opción del menú
    if pregunta == "candidatos a beca":
        #aquí hace una lista con los alumnos que tengan promedio mayor o igual que 8.5
        revisar_diccionario = [nombre for nombre in alumnos if alumnos[nombre] >= 8.5]
        print(revisar_diccionario) #se imprime la lista 
        #preguntamos para que no caiga en bucle
        pregunta = input("Elige una opción entre: agregar/modificar, candidatos a beca, bono de apoyo")

    #tercera opción del menú
    if pregunta == "bono de apoyo":
        #recorre con map y la función hace que se le sume 0.3 y con values() nos arroja los valores
        puntos = map(lambda n: n + 0.3, alumnos.values())
        #recorre el nombre y el promedio
        for nombre, promedio in alumnos.items():

            #haciendo que no se pase de un promedio de 10
            nuevo_promedio = (lambda n: 10 if n + 0.3 > 10 else n + 0.3)(promedio)
            alumnos[nombre] = nuevo_promedio

    #se imprime el diccionario actualizado
        print(f"El diccionario actualizado es: {alumnos}")
        #se vuelve a preguntar para evitar un bucle
        pregunta = input("Elige una opción entre: agregar/modificar, candidatos a beca,bono de apoyo ")

print("Has decidido salir, vuelve pronto :) ")    

```
