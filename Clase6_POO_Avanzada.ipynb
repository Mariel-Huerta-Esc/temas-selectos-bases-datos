## 1. Encapsulamiento y Decoradores (`@property`)

En el diseño de software, no queremos que cualquier parte del programa modifique los atributos de un objeto directamente de forma descontrolada (por ejemplo, asignar un saldo negativo a una cuenta bancaria). 

**Encapsular** significa ocultar el estado interno de un objeto y restringir el acceso directo. En Python:
* Usamos un guion bajo (`_`) para indicar que un atributo es **protegido** (convención para uso interno).
* Usamos doble guion bajo (`__`) para volver un atributo **privado** (*Name Mangling*).
* Usamos el decorador `@property` para crear **Getters** y **Setters** limpios, permitiendo validar datos antes de modificar el estado.

```python
class CuentaBancariaAvanzada:
    def __init__(self, titular, saldo_inicial):
        self.titular = titular
        self.__saldo = saldo_inicial  # Atributo privado

    @property
    def saldo(self):
        """Getter: Permite leer el saldo como si fuera un atributo normal, pero de forma segura."""
        return self.__saldo

    @saldo.setter
    def saldo(self, nuevo_monto):
        """Setter: Modifica el saldo interceptando el valor para validarlo."""
        if nuevo_monto >= 0:
            self.__saldo = nuevo_monto
        else:
            print("Error: El saldo no puede ser negativo.")

# --- Demostración ---
cuenta = CuentaBancariaAvanzada("Carlos", 1000)

# Intentar acceder directamente al atributo privado fallará (Lanzará AttributeError)
# print(cuenta.__saldo)

# Accedemos de manera limpia gracias a @property
print(f"Saldo actual: ${cuenta.saldo}")

# Intentamos modificar el saldo usando el Setter
cuenta.saldo = 1500  # Funciona de forma transparente
print(f"Nuevo saldo: ${cuenta.saldo}")

cuenta.saldo = -500  # Activará nuestra validación de seguridad
print(f"Saldo final: ${cuenta.saldo}")
```

## 2. Herencia y Reutilización de Código

La **Herencia** nos permite crear una nueva clase (clase hija u objeto especializado) a partir de una clase existente (clase padre o base). La clase hija hereda todos los atributos y métodos de la clase padre, pudiendo añadir los suyos propios o modificar el comportamiento existente.

Para invocar al constructor del padre y no duplicar código, utilizamos la función `super()`.

```python
class Empleado:
    """Clase Base (Padre)"""
    def __init__(self, nombre, salario_base):
        self.nombre = nombre
        self.salario_base = salario_base

    def calcular_pago(self):
        return self.salario_base

    def __str__(self):
        return f"{self.nombre} - Salario Base: ${self.salario_base}"


class Gerente(Empleado):
    """Clase Derivada (Hija)"""
    def __init__(self, nombre, salario_base, bono):
        # Invocamos al constructor de Empleado para inicializar nombre y salario_base
        super().__init__(nombre, salario_base)
        self.bono = bono  # Atributo exclusivo de Gerente

    def calcular_pago(self):
        """Sobrescritura de método (Override): Redefinimos cómo calcula el pago un Gerente"""
        return self.salario_base + self.bono

# --- Demostración ---
emp = Empleado("Luis", 2000)
ger = Gerente("Sofia", 3500, 1200)

print(emp, "| Pago Total:", emp.calcular_pago())
print(ger, "| Pago Total:", ger.calcular_pago())
```

## 3. Clases Abstractas y Polimorfismo

### Clases Abstractas
A veces queremos diseñar una clase padre que sirva únicamente como **molde o plantilla**, y no queremos que nadie pueda crear objetos directamente de ella. Esto se conoce como **Clase Abstracta** e implementa métodos abstractos vacíos que las clases hijas están *obligadas* a desarrollar.

### Polimorfismo
El **Polimorfismo** (muchas formas) es la capacidad de tratar a diferentes objetos de la misma manera si comparten la misma interfaz o clase padre. Podemos iterar por una lista de objetos distintos y llamar al mismo método en todos ellos; cada uno responderá según su propia lógica interna.

```python
from abc import ABC, abstractmethod

class Vehiculo(ABC):  # Heredar de ABC convierte a esta clase en Abstracta
    def __init__(self, marca):
        self.marca = marca

    @abstractmethod
    def encender(self):
        """Método abstracto: obliga a las subclases a definir su propio encendido"""
        pass


class Auto(Vehiculo):
    def encender(self):
        return f"El auto {self.marca} ruge con su motor de combustión."

class MotoElectrica(Vehiculo):
    def encender(self):
        return f"La moto {self.marca} se enciende en absoluto silencio."
    

    

# --- Demostración del comportamiento ---

# 1. Si intentamos instanciar la clase abstracta, Python dará un error:
# v = Vehiculo("Generico")  # Descomentar para ver el error TypeError

# 2. Instanciamos las clases concretas
mis_vehiculos = [
    Auto("Mazda"),
    MotoElectrica("Tesla"),
    Auto("Ford")
]

# 3. Aplicamos POLIMORFISMO:
# Tratamos a todos como 'Vehiculo' y llamamos a 'encender()' independientemente de qué tipo exacto sean
for v in mis_vehiculos:
    print(v.encender())
```
