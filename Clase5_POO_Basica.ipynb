## 1. Definición de una clase simple

```python
class Persona:
    """
    Clase que representa a una persona.

    Atributos:
    ----------
    nombre : str
        Nombre de la persona.
    edad : int
        Edad de la persona.
    """

    def __init__(self, nombre, edad):
        """
        Constructor de la clase Persona.

        Parámetros:
        -----------
        nombre : str
            Nombre de la persona.
        edad : int
            Edad de la persona.
        """
        self.nombre = nombre
        self.edad = edad

    def saludar(self):
        """
        Imprime un saludo usando el nombre y la edad de la persona.
        """
        print(f"Hola, soy {self.nombre} y tengo {self.edad} años.")

    def __str__(self):
        """
        Método especial que personaliza cómo se escribe el objeto.
        """
        return f"{self.nombre} ({self.edad} años)"

#---- Cuerpo principal de mi programa---
# Crear un objeto de la clase Persona
p1 = Persona("Juan", 25)

# Llamar al método saludar para mostrar el saludo
p1.saludar()

# Imprimimos solo p1 para ver el resultado de escribir __str__
print(p1)
```

**¿Por qué los métodos usan self?**

En Python, ``self``es una referencia al objeto actual (a la instancia que se creó a partir de la clase). Por ejemplo en el código anterior el método ``saludar`` necesita saber a qué objeto específico se está llamando.

``self`` funciona como puente entre los métodos y los atributos de ese objeto en particular. Sin ``self``, el método no sabría de qué objeto está hablando.

``self.nombre`` significa “el atributo nombre de este objeto específico”.

**¿Por qué se llama self?**

Es convención en Python (como ``this`` en otros lenguajes). Podrías usar otro nombre, pero no es recomendable.

## 2. El "Estado" de un Objeto y Métodos que lo Modifican

Un objeto no es estático; contiene un **estado** (el valor actual de sus atributos en un momento dado). Los métodos no solo sirven para imprimir texto en pantalla, sino para alterar e interactuar con ese estado interno de forma segura.

```python
class CuentaBancaria:
    def __init__(self, titular, saldo_inicial):
        self.titular = titular
        self.saldo = saldo_inicial  # Este es el estado inicial

    def depositar(self, monto):
        """Modifica el estado interno aumentando el saldo."""
        if monto > 0:
            self.saldo += monto
            print(f"Depósito exitoso. Nuevo saldo: ${self.saldo}")
        else:
            print("Error: El monto a depositar debe ser mayor a 0.")

    def __str__(self):
        return f"Cuenta de {self.titular} - Saldo actual: ${self.saldo}"

# --- Demostración ---
cuenta = CuentaBancaria("Ana", 1000)
print(cuenta)

# Modificamos el estado del objeto a través de su método
cuenta.depositar(500)
print(cuenta)
```

## 3. Colaboración entre Objetos

En aplicaciones reales, los objetos no trabajan aislados. Un método de una clase puede recibir como argumento un objeto que pertenezca a otra clase diferente, permitiendo que interactúen entre sí.

```python
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

class CarritoCompras:
    def __init__(self):
        self.productos = []  # Lista contenedora que guardará instancias de la clase Producto

    def agregar_producto(self, producto_objeto):
        """Recibe un objeto de tipo Producto y lo añade a la lista"""
        self.productos.append(producto_objeto)
        print(f"Añadido al carrito: {producto_objeto.nombre}")

    def mostrar_total(self):
        """Accede a los atributos de los objetos almacenados para calcular el total"""
        total = sum(p.precio for p in self.productos)
        print(f"Total a pagar en el carrito: ${total}")

# --- Demostración ---
# 1. Creamos objetos independientes de tipo Producto
p_uno = Producto("Teclado Mecánico", 1200)
p_dos = Producto("Mouse Óptico", 600)

# 2. Creamos el objeto contenedor (Carrito)
mi_carrito = CarritoCompras()

# 3. Pasamos los objetos producto como argumentos al carrito
mi_carrito.agregar_producto(p_uno)
mi_carrito.agregar_producto(p_dos)

# 4. Ejecutamos la lógica de cálculo
mi_carrito.mostrar_total()
```
