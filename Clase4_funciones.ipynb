# Funciones
Una función es un bloque de código reutilizable que solo se ejecuta cuando se le llama. Se define con la palabra clave **def**:

```python
# Sintaxis básica de una función
def nombre_funcion(parámetros):
    # Bloque de código (cuerpo de la función)
    return resultado  # (opcional)
```

**Sintaxis avanzada de funciones en Python**

Python permite definir funciones de forma más expresiva y flexible mediante varias características adicionales:

| Característica                      | Ejemplo de sintaxis                                  | Descripción                                                 |
|------------------------------------|-------------------------------------------------------|-------------------------------------------------------------|
| **Valor por defecto**              | `def saludar(nombre="invitado"):`                    | Si no se pasa argumento, usa el valor predeterminado        |
| **Anotaciones de tipo**            | `def cuadrado(x: int) -> int:`                       | Indica tipos esperados de entrada y salida (no obliga, solo documenta)      |
| **Múltiples argumentos `*args`**   | `def sumar(*numeros):`                               | Recibe varios valores como una tupla                        |
| **Argumentos con nombre `**kwargs`** | `def mostrar_info(**datos):`                        | Recibe argumentos con nombre como un diccionario            |
| **Combinación completa**           | `def func(a: int, b: int = 5, *args, **kwargs) -> int:` | Mezcla parámetros normales, por defecto y flexibles      |

> Estas extensiones hacen que las funciones sean más legibles, reutilizables y seguras. Aunque Python no obliga los tipos, las anotaciones ayudan a documentar y depurar el código.



```python
# Ejemplo
# Definimos una función llamada 'hello' que imprime saludos en diferentes idiomas
def hello():
    print('Hola!')  # Saludo en español
    print('Hello!!!')  # Saludo en inglés
    print('こんにちは')  # Saludo en japonés


# main
hello()  # Llamado a la función para ejecutar su código
```

```python
# Funciones de una calculadora
# Función que suma dos números y devuelve el resultado
def suma(x , y):
    resultado = x + y  # Realiza la suma
    print(f'El resultado de la suma es: {resultado}')  # Muestra el resultado
    return resultado  # Devuelve el resultado

# Función que resta dos números y devuelve el resultado
def resta(x, y):
    resultado = x - y  # Realiza la resta
    print(f'El resultado de la resta es: {resultado}')  # Muestra el resultado
    return resultado  # Devuelve el resultado

res = 0  # Inicializa la variable resultado
res = suma(99, 88)  # Llamado a la función de suma
print(f'Resultado de suma: {res}')
res = resta(5, 9)  # Llamado a la función de resta
print(f'Resultado de resta: {res}')
```

```python
# Ejemplo: Valor por defecto
# Función que recibe un parámetro y lo usa en el mensaje
def hola(nombre = "invitado"):
    print(f'Hola, {nombre}')  # Muestra un saludo con el nombre proporcionado

hola('Han')
hola()
hola(5)  # Llamado a la función con un valor numérico
# print('Hello, ', name)  # Esto generaría un error si 'name' no está definido en el ámbito global
```

```python
# Ejemplo: Anotaciones de tipo
def cuadrado(x: int) -> int:
    return x * x

print(cuadrado(4))
```

```python
# *args: múltiples argumentos posicionales
def sumar(*numeros):
    print(type(numeros))
    return sum(numeros)

print(sumar(1, 2, 3, 4))
```

```python
# **kwargs: múltiples argumentos con nombre
def mostrar_info(**datos):
    print(type(datos))
    for clave, valor in datos.items():
        print(f"{clave}: {valor}")

mostrar_info(nombre="Luis", edad=30, ciudad="CDMX")
```

```python
# Combinación de todas las formas
def funcion_completa(a: int, b: int = 5, *args, **kwargs) -> int:
    print(f"a: {a}, b: {b}, args: {args}, kwargs: {kwargs}")
    return a + b + sum(args)

funcion_completa(2, 3, 4, 5, x=10, y=20)
```

Veamos cómo se reparten los argumentos:

    2 se asigna a a

    3 se asigna a b

    4, 5 se almacenan en args → (4, 5)

    x=10, y=20 no corresponden a ningún parámetro con nombre definido, por lo que se almacenan en kwargs → {'x': 10, 'y': 20}
