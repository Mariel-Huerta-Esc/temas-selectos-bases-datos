# Ejercicio 5:
 Diferencia entre listas y arrays en operaciones

```python
# array 
import numpy as np
x = np.array([1, 2, 3])
x * 3
```

```python
# Usando listas de python 
a = [1, 2, 3]
a * 3
```

Preguntas:
1. ¿Cuál es la diferencia entre los resultados de x * 3 y a * 3?
Con el array solamente multiplica a cada elemento de este por 2 mientras que en la lista repite los elementos de a 3 veces en la lista


2. ¿Por qué ocurre esta diferencia?
Ocurre porque los arrays permiten efectuar operaciones vectorizadas, mientras que una lista no deja hacer operaciones de manera directa, por eso sólo nos repite los elementos de a 3 veces.


3. ¿Cómo podrías modificar el código en la Parte 2 para obtener un comportamiento
similar al de Parte 1 sin transformar la lista en un arreglo?
Pista: En NumPy, los arrays permiten realizar operaciones matemáticas directamente,
mientras que las listas en Python tienen un comportamiento diferente al usar el operador *.

## Lo único que se me ocurrió fue recorrer la lista con un for y hacer que multiplique elemento por elemento... 

```python
#modificando la parte 2: 
a = [1, 2, 3]
resultado = [] #los guarda aquí
contador = 0
for i in a: # recorre "a"
    multiplicacion = i * 3
    resultado.append(multiplicacion) #agrega los números multiplicados en el arreglo
print(resultado)
```
