De tradicional a "Pythónico" (List Comprehensions)

Imagina que tienes una lista con los niveles de varios jugadores en un servidor de
videojuegos: niveles = [5, 12, 18, 24, 31, 45, 60].
Para practicar la sintaxis, analiza primero este ciclo tradicional:

expertos = []

for n in niveles:

if n > 20:

expertos.append(n)

Lo que tienes que hacer:

1. Traduce el ciclo anterior a una sola línea de código utilizando una list
comprehension para generar la lista expertos.

2. Crea una segunda list comprehension llamada niveles_duplicados que
multiplique por 2 todos los niveles de la lista original (sin usar filtros if).

3. Crea una tercera list comprehension llamada etiquetas que use un if-else
interno dentro de la línea: si el nivel es mayor o igual a 30 debe guardar el texto
"Veterano", de lo contrario debe guardar "Novato".

```python
#analisando el siguiente código:
niveles = [5, 12, 18, 24, 31, 45, 60]  
expertos = []
for n in niveles:
    if n > 20:
        expertos.append(n)
print(expertos)
    
```

```python
#reemplazo del código anterior
niveles = [5, 12, 18, 24, 31, 45, 60]  
reemplazo_lineas = [n for n in niveles if n > 20]
print(reemplazo_lineas)
```

```python
#segunda list comprehension
niveles_duplicados = [5, 12, 18, 24, 31, 45, 60] 
multiplicacion = [n * 2 for n in niveles]
print(f"Los números multiplicados por 2 son: {multiplicacion}")
```

```python
#tercera list comprehension
etiquetas = [5, 12, 18, 24, 31, 45, 60]
resultado = ["Veterano" if i >= 30 else "Novato" for i in etiquetas]
print(resultado)


```
