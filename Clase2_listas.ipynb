## **Slicing**

```python
frutas = ["Manzana","Banana","Cereza","Naranja"] #define una lista
print(frutas[1:3])      # Nos devuelve lo que está antes de 3
print(frutas[1:4])
print(frutas[:])    # nos da la lista completa
print(frutas)
```

```python
frutas[0:4:3] #[inicio:fin:salto]
```

```python
# Invertir la lista usando slicing [inicio:fin:paso]
lista_invertida = frutas[::-1]
print(f"Lista original: {frutas}") #imprime la lista normal
print(f"Lista invertida: {lista_invertida}") #imprime la lista invertida
```

```python
frutas_casa = frutas     #No realizamos una copia
frutas_casa[2]="Toronja" 
print(frutas)
print(frutas_casa)
```

### **¿Por qué frutas_casa = frutas no crea una copia?**

En Python, cuando asignas una lista existente a una nueva variable usando el signo =, no estás duplicando los datos. Lo que estás haciendo es crear una nueva "etiqueta" o referencia que apunta al mismo objeto en la memoria de tu computadora.

**Identidad compartida:** Ambas variables, frutas y frutas_casa, apuntan a la misma dirección de memoria.

**Efecto espejo:** Como ambas ven el mismo objeto, cualquier cambio que hagas a través de una variable se reflejará automáticamente en la otra.

**Mutabilidad:** Las listas son objetos mutables, lo que significa que su contenido puede cambiar sin necesidad de crear una lista nueva en memoria.

**Analogía:** Es como si una casa tuviera dos direcciones postales diferentes; no importa a qué dirección envíes un paquete, terminará dentro de la misma casa.

### Cómo solucionarlo (Realizar una copia de verdad)
Si necesitas que frutas_casa sea independiente, debes indicarle a Python explícitamente que cree un objeto nuevo:

- Usando Slicing: ``frutas_casa = frutas[:]``
- Usando el método .copy(): ``frutas_casa = frutas.copy()``

De esta forma, Python reserva un nuevo espacio en memoria y copia los valores uno por uno, rompiendo el vínculo entre ambas variables.

```python
frutas_casa = frutas[:]     #Realizamos una copia
frutas_casa[2]="Toronja"
print(frutas)
print(frutas_casa)
```

```python
frutas_casa = frutas.copy()
frutas_casa[2]="Mango"
print(frutas)
print(frutas_casa)
```

```python
frutas_trabajo = frutas
frutas_trabajo[2]="Toronja"
print(frutas)
print(frutas_trabajo)
```

## El operador de repetición (*) en listas

En Python, cuando usamos el operador de multiplicación * entre una lista y un número entero, no estamos realizando una operación matemática sobre los elementos internos. Lo que ocurre es una concatenación repetida.

Comportamiento: Python toma los elementos de la lista original y los "pega" uno tras otro la cantidad de veces que indiques.

Utilidad: Es extremadamente útil para pre-asignar o inicializar listas con valores predeterminados (como una lista de 100 ceros o valores vacíos) antes de empezar a procesar datos.

```python
[1,2]*6
```

```python
zeros = [1]*100
zeros
```

```python
verduras*2
```

```python
[0]*10
```
