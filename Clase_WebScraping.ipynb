## 1. Introducción
Este notebook está diseñado para ayudarte a aprender los fundamentos del web scraping de forma legal y segura. Utilizaremos bibliotecas como `requests`, `BeautifulSoup` y `pandas`.

El **web scraping** consiste en tomar el contenido que podemos ver en páginas web en HTML y XML de manera automatizada, para transformarlo en bases de datos estructuradas que puedan analizarse fácilmente. 

En lugar de copiar y pegar datos manualmente desde una página web, un programa puede:
- Visitar una URL
- Leer su contenido (normalmente en HTML)
- Extraer partes específicas (como títulos, precios, tablas, imágenes, etc.)
- Guardar o procesar los datos localmente

⚠️ **Importante:** Es importante hacer web scraping de manera ética, legal y responsable. Antes de hacer scraping de cualquier sitio, revisa sus términos de uso y el archivo `robots.txt` para saber qué está permitido, si se prohibe el web scraping en algún sitio, no debe usarse esta técnica.

### **Buenas prácticas**
- Revisa siempre `robots.txt`: Es un archivo de texto que los sitios web colocan en su raíz para indicar qué partes de la página permiten o prohíben indexar/scrapear. Por ejemplo, visita [https://books.toscrape.com/robots.txt](https://books.toscrape.com/robots.txt)
- Establecer un `User-Agent`: Es una cadena de texto en las cabeceras HTTP que le dice al servidor qué navegador o aplicación está haciendo la consulta. Modificarlo ayuda a identificarse correctamente y evitar bloqueos automáticos.
- Simular comportamiento humano: Al hacer peticiones repetitivas, debemos usar pausas (`time.sleep()`) para no saturar ni tirar el servidor del sitio que estamos consultando.
- Priorizar APIs oficiales: Si el sitio cuenta con una API pública para desarrolladores, siempre se debe usar esa vía antes de recurrir al scraping.
- No accedas a información personal o restringida

Es útil cuando necesitamos datos pero no existe una API oficial o los datos no están disponibles en archivos estructurados.

### Riesgos y precauciones del Web Scraping

**Bloqueos y restricciones técnicas**

Muchos sitios detectan y bloquean el tráfico automatizado. Puedes recibir un código 403 o 429, o ser redirigido a páginas vacías.

Algunas contramedidas del sitio:
- Requerir login
- Captchas
- IP bans
- Servir HTML dinámico con JavaScript (más difícil de parsear)

**Problemas legales o éticos**

- Los sitios pueden tener términos de servicio que prohíben explícitamente el scraping.
- Algunos casos famosos han llegado a los tribunales.
- Extraer datos personales sin consentimiento puede violar leyes como el GDPR o la Ley Federal de Protección de Datos Personales en México.

**Lo que NO debes hacer**

- Hacer scraping de redes sociales, bancos, tiendas, plataformas educativas sin permiso.
- Automatizar login de cuentas sin autorización.
- Descargar masivamente archivos o páginas.

**Recomendación profesional**

- Siempre revisa el archivo robots.txt del sitio.
- Por ejemplo: https://example.com/robots.txt
- Si prohíbe /private, no hagas scraping de esa ruta. Aunque no sea legalmente vinculante, muestra la intención del sitio.

### Casos legales famosos de web scraping
1. hiQ Labs vs. LinkedIn (2017–2022) [Enlace:](https://en.wikipedia.org/wiki/HiQ_Labs_v._LinkedIn)
    - Contexto: hiQ Labs era una startup que hacía scraping de perfiles públicos de LinkedIn para ofrecer análisis predictivo de recursos humanos.
        - LinkedIn les envió cartas de "cease and desist" (cese y desista), argumentando que el scraping violaba sus términos de servicio y la Computer Fraud and Abuse Act (CFAA).

    - Resultado: En 2017, un juez prohibió a LinkedIn bloquear a hiQ por considerar que los datos eran públicos y que el scraping no violaba la ley federal.LinkedIn apeló y el caso fue llevado a la Suprema Corte de EE.UU., que devolvió el caso a instancias menores tras un fallo relacionado con el CFAA (caso Van Buren, 2021).En 2022, LinkedIn y hiQ llegaron a un acuerdo fuera de los tribunales.

    - Lecciones: Aun si los datos están "públicos", la legalidad del scraping puede depender del uso y del consentimiento del sitio. Las Términos de Servicio (ToS) pueden no tener fuerza legal inmediata, pero son un primer aviso.

2. Facebook vs. Power Ventures (2010–2017)
    - Contexto: Power Ventures era una plataforma que agregaba datos de redes sociales con el permiso de los usuarios. Usaron credenciales de usuario para iniciar sesión y extraer datos de Facebook.

    - Resultado: Facebook los demandó por violar la CFAA y por evasión técnica de bloqueos (como cambio de IPs). En 2016, la Corte de Apelaciones del 9º Circuito falló a favor de Facebook. En 2017, Power Ventures fue multada y cerró operaciones.

    - Lecciones: Usar cuentas de usuario para scrapear datos de una plataforma cerrada puede violar leyes penales si se evade el bloqueo. El acceso no autorizado y los intentos de evitar medidas de seguridad son factores clave.

3. United States vs. Van Buren (2021) — Aunque no es sobre scraping directo, es clave.
    - Contexto: Un policía accedió a una base de datos oficial para un propósito personal (recibir dinero). Fue acusado de violar la CFAA por "exceder acceso autorizado".

    - Resultado: La Corte Suprema de EE.UU. falló que acceder a un sistema con autorización pero para otro propósito NO viola la CFAA. Esto cambió la interpretación de la ley para muchos casos de scraping.

    - Lecciones: Este fallo limitó la aplicación de la CFAA y abrió margen para casos como el de hiQ Labs, pero no da luz verde al scraping sin reglas.

### ¿Qué es la CFAA (Computer Fraud and Abuse Act)?

La **CFAA (Computer Fraud and Abuse Act)** es una ley federal de EE.UU. aprobada en 1986 para combatir el acceso no autorizado a computadoras, redes y sistemas informáticos.

### ¿Qué prohíbe?
- Acceder a un sistema sin permiso
- Usar una cuenta para fines no autorizados
- Obtener datos confidenciales sin consentimiento
- Dañar, alterar o robar información digital

### ¿Por qué afecta al Web Scraping?
Empresas como **LinkedIn** o **Facebook** han invocado esta ley para impedir el scraping automatizado, especialmente cuando se hace sin respetar términos de uso, evade bloqueos o accede a zonas privadas del sitio.

### Conclusión
Aunque fue creada contra el hacking, la CFAA ha sido usada en disputas de scraping. Por eso:
- Revisa siempre los **términos de servicio**
- Evita scraping de sitios cerrados o que requieren login
- Prioriza fuentes de **datos abiertos o APIs públicas**


## 2. El DOM, IDs y Clases en CSS

Antes de programar, debemos entender cómo lee el navegador una página web.

### 2.1. ¿Qué es el DOM (Document Object Model)?

Cuando un servidor nos envía código HTML, el navegador no lo ve solo como texto; lo organiza en una estructura de árbol jerárquica llamada DOM. En este árbol, cada etiqueta (como ``<html>``, ``<body>``, ``<p>``) es un nodo. Los nodos tienen relaciones de parentesco: padres, hijos y hermanos.

### 2.2 Identificadores en la Web: IDs vs. Clases

Para poder extraer un dato específico, necesitamos saber cómo "apuntarle" usando selectores de CSS:

- ID (id): Es el identificador único de un elemento en toda la página. No puede haber dos elementos con el mismo ID en el mismo documento HTML.

    - Analogía: Es como el número de documento de identidad de una persona.
    - En código/CSS: Se representa con el símbolo # (ej. #link3).

- Clase (class): Es un atributo que se asigna a un grupo de elementos para aplicarles el mismo estilo visual. Un mismo elemento puede tener varias clases y una misma clase puede repetirse cientos de veces en la página.

    - Analogía: Es como el uniforme de un equipo; muchas personas pueden llevarlo puesto al mismo tiempo.
    - En código/CSS: Se representa con un punto . (ej. .product_pod).

## 3. Fundamentos de HTTP y scraping
📌 **¿Qué es HTTP?**

HTTP (HyperText Transfer Protocol) es el protocolo que usamos para comunicarnos con servidores web. Es un protocolo de **petición**-**respuesta** entre un cliente (el navegador o tu script de python) y un servidor web.

🧩 **¿Qué pasa cuando visitas un sitio web?**

El navegador hace una petición HTTP al servidor (por ejemplo: GET /index.html).

El servidor responde con:
- Un código de estado (ej. 200 OK)
- Un cuerpo de respuesta: HTML, XML, JSON, etc.

### Tipos de peticiones HTTP comunes
|Método	|Descripción|
|-------|-----------|
|GET	|Solicita un recurso (página, imagen, etc.)|
|POST	|Envía datos al servidor (formularios, autenticación)|
|HEAD	|Como GET, pero solo obtiene los encabezados|
|PUT, DELETE	|Usados más en APIs|


En web scraping casi siempre usamos GET.

### Códigos de estado HTTP más comunes

|Código|	Significado|	¿Qué hacer?|
|------|---------------|---------------|
|200|	OK|	Todo bien|
|301/302|	Redirección|	Seguir la URL nueva|
|403|	Prohibido (Forbidden)|	Puede estar bloqueado|
|404|	No encontrado|	Recurso no existe|
|429|	Demasiadas peticiones|	¡Haz pausas! Usa sleep()|
|500|	Error interno del servidor|	Espera o intenta más tarde|

## Ejemplo seguro: `books.toscrape.com`
Este sitio fue diseñado para practicar web scraping. Vamos a extraer los títulos y precios de los libros en la primera página del catálogo.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python

# Definir la URL y headers
url = 'https://books.toscrape.com/catalogue/page-1.html'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:140.0) Gecko/20100101 Firefox/140.0'}
```

```python

# Hacer la petición
response = requests.get(url, headers=headers)
if response.status_code == 200:
    print(f'Estado de la petición exitoso.')
    print(response.text[0:40])
else:
    print(f'Estado de la petición: {response.status_code}')
```

```python
# Parsear la respuesta para interpretar los elementos del html
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
# Extraer el header
print(soup.find('title'))
print(soup.title)
print(soup.title.get_text(strip = True))
```

```python
# Título
titulo = soup.find("div", class_="col-sm-8 h1").get_text(strip=True)
print(titulo)
```

```python
# Extraer la información
books = soup.select('.product_pod')
data = []
books
```

```python

for book in books:
    # Nombre del libro
    title = book.h3.a['title']
    #title = book.find('h3').find('a')['title']
    print(title)
    # Precio
    price = book.select_one('.price_color').text
    #price = product.find('p', class_='price_color').text
    print(price)
    # Imagen
    image_tag = book.select_one('.image_container img')
    image_url = "https://books.toscrape.com/" + image_tag['src'].lstrip("../")
    disponibilidad = book.select_one('.availability').text.strip()
    rating = book.select_one('p.star-rating')['class'][1]
    print(image_url)
    data.append({'Title': title, 'Price': price, 'Image': image_url})
    time.sleep(1)  # Buenas prácticas

# Mostrar como DataFrame
df = pd.DataFrame(data)
df.head()
```

## Acceder a las demás páginas


```python
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
headers = {'User-Agent': 'Mozilla/5.0'}

all_data = []

for page_num in range(1, 4):  # Páginas 1 a 3
    print(f"Procesando página {page_num}...")
    url = base_url.format(page_num)
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        books = soup.select('.product_pod')
    except requests.RequestException as e:
        print(f'Error en page {page_num}: {e}')
        continue
    
    for book in books:
        title = book.h3.a['title']
        price = book.select_one('.price_color').text
        image = "https://books.toscrape.com/" + book.select_one('img')['src'].lstrip("../")
        disponibilidad = book.select_one('.availability').text.strip()
        rating = book.select_one('p.star-rating')['class'][1]

        all_data.append({
            'Title': title,
            'Price': price,
            'Disponibilidad': disponibilidad,
            'Rating': rating,
            'Image': image
        })

    time.sleep(2)  # Pausa entre páginas

# Convertir a DataFrame
df_all = pd.DataFrame(all_data)
df_all.head(10)  # Mostrar los primeros 10 resultados
```

## El límite de BeautifulSoup

Las librerías tradicionales como requests y BeautifulSoup realizan peticiones estáticas. Si un sitio web moderno utiliza frameworks de JavaScript (como React, Angular o Vue) para renderizar sus datos en tiempo real, requests sólo descargará un esqueleto de HTML vacío y la información no aparecerá.
Automatización de Navegadores con Selenium

Para solucionar esto, recurrimos a herramientas como Selenium. Selenium no hace una simple petición de texto, sino que levanta una instancia real de un navegador (como Google Chrome) mediante un componente controlador (WebDriver).

Al abrir un navegador real, el JavaScript de la página se ejecuta por completo, permitiéndonos interactuar con el sitio tal como lo haría un humano: haciendo clics en botones, haciendo scroll para cargar más contenido o rellenando formularios de búsqueda de manera programática.


###  Introducción a Selenium

**Selenium** es una herramienta que permite **automatizar navegadores web**. Es muy útil para:

- Hacer pruebas automáticas de aplicaciones web
- Automatizar tareas repetitivas en sitios web (como hacer clic, llenar formularios, etc.)
- Extraer datos de páginas que requieren **interacción humana o cargan con JavaScript**

A diferencia de `requests` + `BeautifulSoup`, Selenium **lanza un navegador real** (como Chrome o Firefox) y lo controla por código. Esto permite simular el comportamiento de un usuario humano.

---

###  ¿Cuándo usar Selenium?

| Caso de uso                              | ¿Usar Selenium? |
|------------------------------------------|------------------|
| Página con contenido estático (HTML)     |  Mejor usar requests + BeautifulSoup |
| Página con JavaScript que genera datos   |  Sí, Selenium es útil aquí |
| Necesitas hacer clics, login, scroll     |  Selenium puede hacerlo |

---



```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 1. Inicializar el navegador de forma automática
print("Abriendo el navegador...")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

try:
    # 2. Visitar la página web de frases
    url = "https://quotes.toscrape.com"
    driver.get(url)
    print(f"Navegando en: {driver.title}\n")

    # Vamos a scrapear las 2 primeras páginas haciendo "clic" humano
    for pagina in range(1, 3):
        print(f"--- EXTRAYENDO DATOS DE LA PÁGINA {pagina} ---")
        
        # Buscar todas las cajas de las citas usando su Clase CSS (.quote)
        citas = driver.find_elements(By.CLASS_NAME, "quote")
        
        for cita in citas:
            # Buscar el texto y el autor dentro de cada caja usando Clases CSS
            texto = cita.find_element(By.CLASS_NAME, "text").text
            autor = cita.find_element(By.CLASS_NAME, "author").text
            print(f"{texto} — {autor}")
        
        print("-" * 40)
        
        # 3. ¡La magia de Selenium! Simular un clic en el botón "Next" para avanzar
        if pagina < 2:  # Solo hacemos clic si no es la última página que queremos
            try:
                print("Simulando clic en el botón 'Next'...")
                # Buscamos el botón por su clase CSS (.next) y luego la etiqueta 'a'
                boton_siguiente = driver.find_element(By.CLASS_NAME, "next").find_element(By.TAG_NAME, "a")
                boton_siguiente.click()
                
                # Buena práctica: Esperar 2 segundos para que cargue la nueva página antes de seguir
                time.sleep(2) 
            except Exception as e:
                print("Ya no hay más páginas disponibles.", e)
                break

finally:
    # 4. Cerrar el navegador pase lo que pase para no dejar procesos colgados
    print("\nCerrando el navegador...")
    driver.quit()
    print("¡Proceso finalizado!")
```
