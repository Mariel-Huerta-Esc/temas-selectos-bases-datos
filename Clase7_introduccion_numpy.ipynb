### Comparación: Listas vs. NumPy

```python
arreglo = np.array([1, 4, 2, 5, 3])
print(arreglo)
```

```python
# Crear un arreglo a partir de una lista
#arreglo = np.array([10]*10)
arreglo = np.array([1, 4, 2, 5, 3],dtype='int16')
print(arreglo)
for i in arreglo:
    print(type(i))


```

```python
arreglo = np.array([1, 4, 2.5, 5, 3],dtype='float32')
print(arreglo)
for i in arreglo:
    print(type(i))
```

```python
arreglo = np.array([[1,0],[0,1]],dtype='float32')
print(arreglo)
print(type(arreglo))
for i in arreglo:
    for j in i:
        print(j,":",type(j))
```

### 1. Velocidad y Eficiencia

```python
import time

# Con listas de Python
lista = list(range(1_000_000))
start_time = time.time()
lista_squared = [x**2 for x in lista]
print("Tiempo con listas:", time.time() - start_time)

# Con NumPy
array = np.arange(1_000_000)
start_time = time.time()
array_squared = array**2
print("Tiempo con NumPy:", time.time() - start_time)
```

Consistencia en los Tipos de Datos

```python
lista = [1, "hola", 3.14, True]  # Mezcla de tipos en listas

array = np.array([1, 2, 3, 4])  # NumPy convierte todo a un solo tipo (int)
array.dtype
```

Operaciones Matemáticas Vectorizadas

```python
lista = [1, 2, 3, 4]
lista_doble = [x * 2 for x in lista]  # Necesita un bucle

array = np.array([1, 2, 3, 4])
array_doble = array * 2  # Operación directa y más rápida
```

Funcionalidad Adicional

```python
array = np.array([1, 2, 3, 4])
print("Raíz cuadrada:", np.sqrt(array))
print("Suma acumulada:", np.cumsum(array))
print("Media:", np.mean(array))
```

Arreglos Multidimensionales

```python
# Crear un array 1D
arr1 = np.array([1, 2, 3, 4])

# Crear una matriz 2D
arr2 = np.array([[1, 2, 3], [4, 5, 6]],dtype='int8')

# Crear un array 3D (cubo de datos)
arr3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]],dtype='float32')

print(arr1, "\n\n", arr2, "\n\n", arr3)
```

Propiedades de los arrays:

```python
print("Forma:", arr1.shape)  # (filas, columnas)
print("Dimensiones:", arr1.ndim)
print("Tamaño total:", arr1.size)
print("Tipo de datos:", arr1.dtype)
print("Itemsize:",arr1.itemsize, "bytes")
print("nbytes:", arr1.nbytes,"bytes")
```

```python
print("Forma:", arr2.shape)  # (filas, columnas)
print("Dimensiones:", arr2.ndim)
print("Tamaño total:", arr2.size)
print("Tipo de datos:", arr2.dtype)
print("Itemsize:",arr2.itemsize, "bytes")
print("nbytes:", arr2.nbytes,"bytes")
```

```python
print("Forma:", arr3.shape)  # (filas, columnas)
print("Dimensiones:", arr3.ndim)
print("Tamaño total:", arr3.size)
print("Tipo de datos:", arr3.dtype)
print("Itemsize:",arr3.itemsize, "bytes")
print("nbytes:", arr3.nbytes,"bytes")
```

Operaciones aritméticas con arrays:

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b)
print("División:", a / b)
print("Potencia:", a**2)
```

Funciones matemáticas comunes:

```python
x = np.array([0, np.pi / 2, np.pi])

print("Seno:", np.sin(x))
print("Coseno:", np.cos(x))
print("Raíz cuadrada:", np.sqrt([4, 9, 16]))
print("Logaritmo:", np.log([1, np.e, np.e**2]))
```

Caso práctico: Generar datos de salarios de 100 empleados con distribución normal:

```python
salarios = np.random.normal(loc=50000, scale=10000, size=100)
print("Salario promedio:", np.mean(salarios))
print("Desviación estándar:", np.std(salarios))
```
