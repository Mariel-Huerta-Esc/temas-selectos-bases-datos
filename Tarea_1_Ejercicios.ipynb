 Registro de videojuegos (Tuplas y Diccionarios)
Un estudio de videojuegos quiere organizar los datos de su última trilogía utilizando
estructuras inmutables para los datos fijos y diccionarios para su organización.
1. Crea un diccionario llamado catalogo donde las llaves sean los nombres de 3
videojuegos y los valores sean tuplas que contengan tres datos: (año de
lanzamiento, calificación, ventas en millones).
2. Diseña una función llamada mostrar_reporte(catalogo) que recorra el
diccionario utilizando un ciclo for e imprima la información de cada juego
formateada limpiamente usando f-strings.
3. Agrega una estructura condicional (if-else) dentro de la función para que, si la
calificación del juego (segundo elemento de la tupla) es mayor o igual a 90, se
imprima al lado del título la leyenda: "¡Una obra maestra!".

```python
#diccionario 
catalogo = ({
    "The Walking Dead": ["2012", "90", "50 millones"],
    "The Last Of Us" : ["2013", "95", "37 millones"],
    "Call Of Duty" : ["2003", "85", "425 millones"]
})

#funcion
def mostrar_reporte(catalogo):

#for
    for juego, datos in catalogo.items():
        print(f"Juego: {juego}")
        print(f"Año: {datos[0]}")
        print(f"Calificación: {datos[1]}") #imprimiendo
        print(f"Ventas: {datos[2]}")





        if int(datos[1]) >= 90:
         print(f"¡{juego}, una obra maestra!")
        else:
         print(f"{juego} NO tiene una calificación mayor a 90")

      
mostrar_reporte(catalogo) #muestra
```

# Ejercicio 2:

El festival de música (Operaciones con Conjuntos)

La universidad está organizando un festival de música de dos días. Tienes un
conjunto (set) con los nombres de los alumnos que asistirán el Viernes y otro
conjunto con los que asistirán el Sábado (asegúrate de incluir algunos nombres
repetidos en ambos días para simular alumnos que van los dos días).
Escribe un programa que, utilizando operaciones lógicas de conjuntos, determine de
forma automática y muestre en pantalla:

Asistencia Total: ¿Quiénes irán al festival en total (sin repetir nombres)?
(Unión).

• Fieles al Festival: ¿Qué alumnos asistirán ambos días? (Intersección).

• Solo el Viernes: ¿Qué alumnos van el viernes pero decidieron no ir el sábado?
(Diferencia).

```python
nombres_alumnos_viernes = {"Leonardo", "Ana", "Carlos", "Fabián", "Gustavo"}
nombres_alumnos_sabado = {"Leonardo", "Karen", "Carlos", "Juan", "Gustavo"}



#¿Quiénes irán al festival en total 
sin_repetir_nombres = nombres_alumnos_viernes | nombres_alumnos_sabado

# ¿Quiénes asistirán los dos días al festival?
asistiran_dos_dias = nombres_alumnos_viernes & nombres_alumnos_sabado

#¿Qué alumnos van el viernes pero decidieron no ir el sábado?
iran_el_sabado = nombres_alumnos_viernes - nombres_alumnos_sabado

print(f"Unión: {sin_repetir_nombres}")
print(f"Intersección: {asistiran_dos_dias}")
print(f"Diferencia: {iran_el_sabado}")
```

```python
#analizando el siguiente código:
niveles = [5, 12, 18, 24, 31, 45, 60]  
expertos = []
for n in niveles:
    if n > 20:
        expertos.append(n)
print(expertos)
    
```

```python
#reemplazo del código anterior
niveles = [5, 12, 18, 24, 31, 45, 60]  
reemplazo_lineas = [n for n in niveles if n > 20]
print(reemplazo_lineas)
```

```python
#segunda list comprehension
niveles_duplicados = [5, 12, 18, 24, 31, 45, 60] 
multiplicacion = [n * 2 for n in niveles]
print(f"Los números multiplicados por 2 son: {multiplicacion}")
```

# Ejercicio 4.

 El radar de velocidad y descuentos (Funciones Lambda)

Tienes una lista con los precios originales de varios artículos en una tienda de
tecnología: precios = [100, 250, 80, 1500, 420, 60].

Lo que tienes que hacer es:

1. Crea una función lambda llamada calcular_descuento que reciba un precio y
calcule su valor final tras aplicarle un 15% de descuento (es decir,
multiplicarlo por 0.85).

2. Investiga y utiliza la función nativa map() junto con tu función lambda para
aplicar el descuento a toda la lista y generar una nueva lista llamada
precios_buenos. Recuerda convertir el resultado a lista usando list().

3. Crea otra función lambda que sirva para filtrar y, utilizándola junto con la
función filter(), extrae únicamente los precios de la lista precios_buenos que
sean menores a $300.

4. Imprime las listas resultantes para comprobar tus operaciones.

```python
#lista de precios
precios = [100, 250, 80, 1500, 420, 60]
calcular_descuento = lambda precios: precios * 0.85 #calcula descuento, lambda nos ayuda a crear una función exprés
print(calcular_descuento(250))
```

```python
#map sirve para aplicar una función a cada elemento de una lista
precios = [100, 250, 80, 1500, 420, 60]

#aplicando map()
descuento = map(lambda n: n * 0.85, precios) #funcion lambda
precios_buenos = list(descuento) #crea lista con los precios modificados
print(precios_buenos)



"""Aquí tal vez pude haber hecho el código más corto pero fue la única manera en la que se me ocurrió hacerlo"""
```

```python
#funcion lambda para filtrar
filtrar = lambda n: n < 300
precios_filtrados = list(filter(filtrar, precios_buenos))
print(precios_filtrados)
```

```python
 #listas resultantes

print(calcular_descuento(250)) #aquí sólo calculé un solo valor de la lista 
print(f"Los precios con descuentos son: {precios_buenos}")
print(f"los precios filtrados son: {precios_filtrados}")
```
