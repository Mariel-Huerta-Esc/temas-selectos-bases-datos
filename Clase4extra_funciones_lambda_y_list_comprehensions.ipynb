# Funciones lambda y list comprehensions en Python

## 1. Funciones lambda
Las funciones `lambda` son funciones anónimas, útiles para expresiones cortas.

**Sintaxis básica:**
```python
lambda argumentos: expresión
```

**Ejemplo:**

````python
suma = lambda a, b: a + b
print(suma(3, 4))  # 7
````

Se pueden usar en funciones como `map()`, `filter()`, etc.

**Ejemplo de uso con `filter()`:**

````python
numeros = [1, 2, 3, 4, 5, 6]
pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)  # [2, 4, 6]
````

## 2. List comprehensions
Las comprensiones de listas permiten crear listas de manera más concisa.

**Sintaxis básica:**
```python
[ expresión for elemento in iterable if condición ]
```

**Ejemplo sin condición:**

````python
numeros = [1, 2, 3, 4, 5]
cuadrados = [x**2 for x in numeros]
print(cuadrados)  # [1, 4, 9, 16, 25]
````

**Ejemplo con condición:**

````python
pares = [x for x in numeros if x % 2 != 0]
print(pares)  # [2, 4]
````

## 3. Ejemplo práctico: eliminar todos los elementos iguales a 4 en una lista
(resuelve la duda de la clase pasada)

````python
lista = [2, 6, 4, 3, 4, 9, 4]
sin_cuatro = [x for x in lista if x != 4]
print(sin_cuatro)  # [2, 6, 3, 9]
````
