**Identado**

El identado (también llamado sangría o tabulación) en Python no es opcional, sino que es parte de la sintaxis del lenguaje. Sirve para indicar qué instrucciones pertenecen a un bloque de código, como los que siguen a un if, for, while o una función.

- El identado se hace generalmente con 4 espacios.
- Todas las líneas que forman parte del mismo bloque deben tener el mismo nivel de sangría.
- Si no se indenta correctamente, Python mostrará un error de sintaxis.

El identado permite que el código sea más claro, legible y estructurado, y es esencial para que el programa se ejecute correctamente.

**¿Qué es la sentencia pass en Python?**

La palabra clave pass se utiliza cuando necesitas escribir una estructura de control (como if, for, o una función), pero aún no quieres escribir su contenido. Es una instrucción nula que le dice a Python: “no hagas nada aquí por ahora”.

Es útil cuando estás construyendo el esqueleto de tu código y quieres evitar errores de sintaxis mientras lo completas más adelante.

## 2. Bucles
### For

```python
# Sintaxis del ciclo for
iterable=[1,2,3,4]
for variable in iterable:
    # Código que se repite por cada elemento del iterable
    pass    # la sentencia pass indica que aún no se ha escrito código
```

**¿Qué es un iterable?**

Un iterable es una secuencia de elementos como una lista, tupla, cadena de texto, o un rango de números.
El for toma cada elemento uno por uno y ejecuta el bloque de código correspondiente.

```python

# 2a. For con listas
nombres = ["Ana", "Luis", "Marta"]
for nombre in nombres:
    print(nombre)
```

```python
# Ejemplo 3
estudiantes_python = ["Juan","María","Ana","Diego"]

for estudiante in estudiantes_python:
    print(estudiante)
print("Fin de la lista")
```

```python
# Ejemplo 4
numeros = [10,25,12,36]

for numero in numeros:
    cuadrado = numero**2
    print(f"El cuadrado de {numero} es {cuadrado}")
```

**¿Qué es range() en Python?**

La función range() se usa comúnmente con for para generar una secuencia de números enteros.

Tiene tres formas principales:

```python
inicio = 2
fin = 20
paso = 2
range(fin)               # Desde 0 hasta fin - 1
range(inicio, fin)       # Desde inicio hasta fin - 1
range(inicio, fin, paso) # Desde inicio hasta fin - 1, avanzando de paso en paso
```

```python
for i in range(5,10,2):
    print(i)
```

- range() no genera una lista visible por sí sola, pero se puede recorrer con for.
- Es muy útil para repetir acciones un número específico de veces.

```python
rango = range(5)
print(rango)
mi_rango = list(rango)
print(mi_rango)
```

```python
# 2b. For con range
for i in range(5):
    print("Hola")
    print(f"Iteración {i}")
```

```python
# Ejemplo 5
a = 2
b = 3
suma = a + b
print(f"El resultado de la suma es {suma}")

potencia = 3
resultado_potencia = 1
for i in range(potencia):
    resultado_potencia = resultado_potencia*suma
print(f"La suma elevada al cubo es {resultado_potencia}")
```

### While

### **Sentencias break y continue**

Las sentencias break y continue se usan dentro de bucles (for o while) para controlar el flujo de repetición:

- **break**: termina el ciclo por completo en el momento en que se ejecuta, incluso si aún no se han recorrido todos los elementos.

- **continue**: salta a la siguiente iteración del ciclo, omitiendo las líneas siguientes del bloque actual.

Son útiles cuando necesitas detener o saltarte partes específicas de un ciclo según alguna condición.

```python
# 2d. Break y continue
# Break
for i in range(10):
    if i == 5:
        break
    print(i)

# Continue
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

### Ejemplos integradores

```python
# Ejemplo: Validación de usuario y contraseña usando break y continue

print("Sistema de acceso seguro\n---------------------------")

while True:
    name = input("¿Quién eres? ").strip()  # Elimina espacios en blanco
    
    if name != 'Magali':
        print("Nombre incorrecto. Intenta de nuevo.\n")
        continue  # Repite el bucle
    
    # Nombre correcto, se solicita la contraseña
    password = input('Hola Magali. ¿Cuál es tu contraseña? (Es un tipo de pez): ').strip()
    
    if password == 'pez espada':
        break  # Sale del bucle si la contraseña es correcta
    else:
        print("Contraseña incorrecta. Vuelve a ingresar tu nombre.\n")

print("\n✅ Acceso concedido.")
```

```python
# Ejemplo: Conteo de edades
edades = [12, 45, 67, 34, 15, 90, 23, 17]
menores = 0
adultos = 0
adultos_mayores = 0

for edad in edades:
    if edad < 18:
        menores += 1
    elif edad < 65:
        adultos += 1
    else:
        adultos_mayores += 1

print(f"Menores de edad: {menores}")
print(f"Adultos: {adultos}")
print(f"Adultos mayores: {adultos_mayores}")
```

```python
# Ejemplo: Ingreso promedio
ingresos = [12000, -1, 3400, 5000, 7600, -2, 3100]
suma = 0
conteo = 0

for ingreso in ingresos:
    if ingreso < 0:
        continue
    suma += ingreso
    conteo += 1

if conteo > 0:
    promedio = suma / conteo
    print(f"Ingreso promedio: {promedio}")
else:
    print("No hay ingresos válidos")
```
