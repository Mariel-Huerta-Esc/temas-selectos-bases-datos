 El radar de velocidad y descuentos (Funciones Lambda)

Tienes una lista con los precios originales de varios artículos en una tienda de
tecnología: precios = [100, 250, 80, 1500, 420, 60].

Lo que tienes que hacer es:

1. Crea una función lambda llamada calcular_descuento que reciba un precio y
calcule su valor final tras aplicarle un 15% de descuento (es decir,
multiplicarlo por 0.85).

2. Investiga y utiliza la función nativa map() junto con tu función lambda para
aplicar el descuento a toda la lista y generar una nueva lista llamada
precios_buenos. Recuerda convertir el resultado a lista usando list().

3. Crea otra función lambda que sirva para filtrar y, utilizándola junto con la
función filter(), extrae únicamente los precios de la lista precios_buenos que
sean menores a $300.

4. Imprime las listas resultantes para comprobar tus operaciones.

```python
#lista de precios
precios = [100, 250, 80, 1500, 420, 60]
calcular_descuento = lambda precios: precios * 0.85 #calcula descuento, lambda nos ayuda a crear una función exprés
print(calcular_descuento(250))
```

```python
#map sirve para aplicar una función a cada elemento de una lista
precios = [100, 250, 80, 1500, 420, 60]

#aplicando map()
descuento = map(lambda n: n * 0.85, precios) #funcion lambda
precios_buenos = list(descuento) #crea lista con los precios modificados
print(precios_buenos)

"""Aquí tal vez pude haber hecho el código más corto pero fue la única manera en la que se me ocurrió hacerlo"""
```

```python
#funcion lambda para filtrar
filtrar = lambda n: n < 300
precios_filtrados = list(filter(filtrar, precios_buenos))
print(precios_filtrados)
```
