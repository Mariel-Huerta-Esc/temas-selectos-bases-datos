## **Reglas para nombrar variables en Python:**

1. El nombre debe comenzar con una letra o un guion bajo (_).
    - *Ejemplo válido: ``_variable`` o ``nombreVariable``.*
2.  No puede comenzar con un número.
    - *Ejemplo no válido: ``3variable``.*
3.  Solo se permiten letras, números y guion bajo.
    - *Ejemplo válido: ``mi_variable``, ``edad2``.*
    - *Ejemplo no válido: ``mi-variable`` o ``nombre*``.*
4. Los nombres de las variables son sensibles a mayúsculas y minúsculas.
    - ``Variable`` y ``variable`` son diferentes.
5. No se pueden usar palabras reservadas de Python como nombres de variables (por ejemplo, ``if``, ``while``, ``class``).

### **Operaciones fundamentales**

**Operaciones aritméticas:**

Python soporta las operaciones matemáticas básicas:


```python
# Declaramos las variables a y b, y les asignamos el valor 10 y 3
a = 10
b = 3

# Suma
print(a + b)  # Salida: 13
```

```python
# Resta
print(a - b)  # Salida: 7
```

```python
# Multiplicación
print(a * b)  # Salida: 30
```

```python
# División con decimales
print(a / b)  # Salida: 3.3333333333333335
```

```python
#División entera
print(a//b) # Salida: 3
```

**Operaciones de módulo (resto de división):**


```python
# Módulo
print(a % b)  # Salida: 1
```

**Potencias:**


```python
# Exponentes
print(a ** b)  # Salida: 1000 (10^3)
```

**Operadores lógicos:**

Los operadores lógicos se usan para combinar valores booleanos:

1. ``and``: Retorna ``True`` si ambos valores son ``True``.


2. ``or``: Retorna ``True`` si al menos uno de los valores es ``True``.


3. ``not``: Invierte el valor lógico (de ``True`` a ``False`` y viceversa).

**Operadores Relacionales:**
x = True, 
y = False

|Operador   |Nombre    |Ejemplo|
|-----------|----------|---------------|
|==	        |Igual	    |x == y = False|
|!=	        |Distinto	|x != y = True|
|>	        |Mayor	    |x > y = False|
|&lt;	    |Menor	    |x &lt; y = True
|>=	        |Mayor o igual	|x >= y = False|
|&lt;=	    |Menor o igual	|x &lt;= y = True|

**Ejemplo:**

```python
x = True
y = False

# AND
print(x and y)  # Salida: False (ambos tienen que ser True)

# OR
print(x or y)   # Salida: True (uno de los dos es True)

# NOT
print(not x)    # Salida: False (invierte el valor de x)
```

**Ejemplo Integrado:**

```python

# Usando variables, tipos de datos y operaciones
precio = 100  # int
descuento = 0.2  # float Tiene un 20% de descuento

precio_final = precio * (1 - descuento)
print("El precio final es:", precio_final)  # Salida: El precio final es: 80.0

# Comprobando si es mayor a un valor
print(precio_final > 50)  # Salida: True
```

### Ejercicio: Calculadora de descuento e impuesto en el precio final de un producto

**Instrucciones:**

Solicita al usuario que ingrese la siguiente información utilizando input():
- Precio original del producto
- Porcentaje de descuento aplicado
- Porcentaje de impuesto aplicado

**Objetivo: Calcular y mostrar**
- El monto del descuento
- El precio después del descuento
- El monto del impuesto
- El precio final del producto (considerando descuento e impuesto)
- Un resumen claro y ordenado de los resultados

**Datos de entrada esperados**
- Precio original (número con o sin decimales)
- Porcentaje de descuento (entero o decimal)
- Porcentaje de impuesto (entero o decimal)

**Ejemplo de ejecución**

Supongamos que el usuario ingresa:

    Precio original: 1000
    Porcentaje de descuento: 20
    Porcentaje de impuesto: 16

El programa debería mostrar algo como:

- Precio original: $1000.00
- Descuento aplicado (20%): -$200.00
- Precio después del descuento: $800.00
- Impuesto aplicado (16%): +$128.00
- Precio final del producto: $928.00

**Nota:** La función ``round(precio_final,2)`` nos permite redondear a dos decimales el precio final

```python
precio_original = int(input("Ingresa el precio original del producto: "))
porcentaje_descuento_aplicado = float(input("Ingresa el porcentaje de descuento aplicado: "))
porcentaje_impuesto_aplicado = float(input("Ingresa el porcentaje del impuesto aplicado: "))

#monto del descuento
monto_descuento = (precio_original * porcentaje_descuento_aplicado) / 100
#precio después del descuento
precio_con_descuento = precio_original - monto_descuento
#monto del impuesto
monto_impuesto = (precio_con_descuento * 16) / 100
#precio final del producto considerando descuento e impuesto
precio_final = precio_con_descuento + monto_impuesto 

#imprime los datos que le dió el usuario
print(f"el precio original es {precio_original}")
print(f"porcentaje de descuento aplicado {porcentaje_descuento_aplicado}")
print(f"porcentaje_impuesto {porcentaje_impuesto_aplicado}")

#mostrando datos:
print(f"El monto del descuento es: {monto_descuento}")
print(f"El precio después del descuento es: {precio_con_descuento}")
print(f"El monto del impuesto es: {monto_impuesto}")
print(f"El precio final del producto es: {precio_final}") #H


```
