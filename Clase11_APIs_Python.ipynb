## ¿Qué es una API?

Una API (Application Programming Interface) es un conjunto de reglas que permiten a dos aplicaciones comunicarse entre sí. En el contexto de ciencia de datos, muchas APIs públicas permiten acceder a información útil (clima, videos, películas, redes sociales, etc.) en formato JSON.

### Conceptos básicos:
- **Endpoint (punto final)**: URL que representa un recurso de la API.
- **Métodos HTTP**:
    - `GET`: para obtener información del punto final
    - `POST`: crear/realizar datos en ese punto final
    - `PUT`: se utiliza para actualizar los datos que creamos.
    - `DELETE`: se utiliza para eliminar los datos que hemos creado.
- **Respuesta**: normalmente en formato **JSON**.

Posteriormente se especifican los headers y payload. En este ejemplo no los requerimos por lo tanto no los hacemos.

Los **headers** son metadatos que acompañan la petición. Sirven para dar instrucciones o información adicional al servidor.

Ejemplos comunes de headers:

- Content-Type: indica el tipo de contenido que estás enviando (application/json, text/plain, etc.).
- Authorization: se usa para enviar tokens de autenticación (como JWT o API keys).
- User-Agent: identifica el cliente que hace la petición (por ejemplo, un navegador o una app).
- Accept: indica qué formatos de respuesta acepta el cliente.
Ejemplo:
> headers = {
>   "Content-Type": "application/json",
>   "Authorization": "Bearer tu_token_aqui"
> }

El **payload** es el contenido o datos que se envían al servidor, especialmente en peticiones POST, PUT o PATCH. Es lo que quieres "enviar" realmente, como datos de un formulario, parámetros de búsqueda, etc.
> payload = {
>    "nombre": "Juan",
>    "email": "juan@ejemplo.com"
> }

El siguiente paso es enviar nuestra petición usando la función `request` del módulo `requests` y utilizamos GET, así la función sabe que esperamos datos de regreso.

Una petición que use su headers y payload se vería de la siguiente manera:

**Convertir nuestros datos a un dataframe**

Cuando trabajamos con datos en formato JSON u otras estructuras similares, es importante elegir el método adecuado para convertirlos en un DataFrame. La función `pd.DataFrame(data)` se utiliza cuando los datos tienen una estructura tabular simple, como una lista de diccionarios sin anidamiento. En cambio, `pd.json_normalize(data)` es útil cuando los datos están anidados (por ejemplo, diccionarios dentro de otros diccionarios), ya que permite "aplanar" esa estructura jerárquica y convertirla en un formato plano donde las claves anidadas se transforman en columnas del DataFrame.

**Ejemplos de `pd.DataFrame(data)` y `pd.json_normalize(data)`**

````python
# Ejemplos de `pd.DataFrame(data)` y `pd.json_normalize(data)`
from pandas import json_normalize

# Ejemplo 1: Datos planos
data_plano = [
    {"nombre": "Juan", 
     "edad": 30},
    {"nombre": "Ana", 
     "edad": 25}
]

df_plano = pd.DataFrame(data_plano)
print("DataFrame con pd.DataFrame():")
display(df_plano)
````

````python

# Ejemplo 2: Datos anidados
data_anidado = [
    {
        "nombre": "Juan",
        "info": {
            "edad": 30,
            "ciudad": "CDMX"
        }
    },
    {
        "nombre": "Ana",
        "info": {
            "edad": 25,
            "ciudad": "Guadalajara"
        }
    }
]

df_anidado = json_normalize(data_anidado)
print("\nDataFrame con json_normalize():")
display(df_anidado)
````

## Ejemplo 2: Studio Ghibli API (Películas)

````python

url = "https://ghibliapi.vercel.app/api/films"
response = requests.get(url)
peliculas = response.json()
peliculas
````

La siguiente celda nos ayuda porque a veces falla la api o ya no da acceso a sus datos entonces los copiaremos en un json

````python
import json

# Abrimos y cargamos el archivo JSON local
with open("ghibli.json", "r", encoding="utf-8") as archivo:
    data = json.load(archivo)

# Convertimos en DataFrame
peliculas = pd.json_normalize(data)

# Mostramos las primeras filas
peliculas.head()
````

````python
peliculas.to_csv("ghibli.csv")
````

### ¿Cuál es la película más larga?

````python

peliculas.sort_values("running_time", ascending=False).head(5)
````

### ¿Cuántas películas tiene cada director?

````python

peliculas["director"].value_counts()
````

````python

# URL de la Dog API (no requiere autenticación para este endpoint)
url = "https://api.thedogapi.com/v1/breeds"

# Hacemos la petición
response = requests.get(url)

# Verificamos si fue exitosa
if response.status_code == 200:
    data = response.json()
    df = pd.json_normalize(data)
    display(df.head())  # Muestra las primeras filas
else:
    print("Error al obtener datos:", response.status_code)
````
