## 1. Importación de módulos y funciones en Python

Python permite organizar el código en módulos: archivos .py que contienen funciones, clases o variables.

- Cuando quieres reutilizar código de otro módulo, lo importas.

### import nombre_modulo

```python
#Ejemplo 1

import random   #Importamos el módulo random

print(f"Número entero aleatorio: {random.randint(4,10)}")

random.seed(2)  # Semilla, al activarla siempre tendremos la misma secuencia de números
for i in range(5):
    print(random.randint(1, 10))
```

```python
# Ejemplo 2
import math

print(math.sqrt(16))  # Usamos la función sqrt del módulo math
print(math.pi)        # Accedemos a la constante pi
```

```python
# Ejemplo 3
import os
print("Directorio actual:", os.getcwd())
```

```python
# Ejemplo 4
import sys
print("Versión de Python:", sys.version)
```

El módulo completo se importa y para usarlo debes anteponer su nombre (math.sqrt, math.pi).

### from nombre_modulo import nombre_funcion

```python
# Ejemplo 5
from random import randint

for i in range(5):
    print(randint(1, 10))
```

```python
# Ejemplo 6
from math import sqrt

print(sqrt(25))  # Ya no necesitas anteponer "math."
```

- Solo importas lo que necesitas (en este caso, sqrt).

- Puedes importar varias funciones separadas por coma:

```python
from math import sqrt, pi
```

### from nombre_modulo import *
(Importación global)

```python
# Ejemplo 7
from math import *

print(sqrt(36))
print(pi)
```

Esto importa todo el contenido del módulo.

⚠️ Cuidado: puede sobrescribir funciones con el mismo nombre que ya tengas en tu script (¡mejor evitarlo en proyectos grandes!).

### Renombrar al importar (as)

```python
# Ejemplo 8
import math as m

print(m.sqrt(49))
```

Cambia el nombre del módulo en tu código a un alias más corto o claro.

### Importar módulos propios

```python
# Ejemplo 9

# Dejar aquí
import mi_modulo

nombre = input("¿Cómo te llamas?")
mi_modulo.saludar(nombre)
print(f"Tu nombre tiene {mi_modulo.contar_letras(nombre)} letras")
```

```python
# Directamente la función
from mi_modulo import saludar
saludar("Ana")
```

En resumen

- Usa import si necesitas varias funciones del módulo.
- Usa from ... import ... si solo quieres una o dos funciones.
- Evita usar from ... import * en programas grandes: puede causar errores difíciles de detectar.

## 2. Manejo de errores con try, except, else, finally
En Python, los errores se llaman excepciones. Si no se manejan, pueden detener tu programa.

- Para evitarlo, se usa el bloque try–except.

### Estructura básica

```python
# Ejemplo 2.1
try:
    # Código que podría causar un error
    resultado = 10 / 0
except ZeroDivisionError:
    # Código que se ejecuta si ocurre ZeroDivisionError
    print("¡Error! No puedes dividir entre cero.")
```

El bloque except captura el error y permite manejarlo sin detener el programa.

### Múltiples tipos de error
Puedes capturar distintos tipos de errores:

```python
a = int(input("Ingresa un número: "))
```

```python
# Ejemplo 2.2
try:
    a = int(input("Ingresa un número: "))
except ValueError:
    print("¡Error! Debes ingresar un número.")
```

```python
# Ejemplo 2.3
try:
    a = int(input("Dame un número: "))
    b = int(input("Otro número: "))
    resultado = a / b
    print("El resultado es:", resultado)
except ZeroDivisionError:
    print("¡Error! No puedes dividir entre cero.")
except ValueError:
    print("¡Error! Ingresa solo números.")
except Exception as e:
    print("Ocurrió un error inesperado:", e)
```

```python
# Ejemplo 2.4
try:
    resultado = 10 / 0
except (ZeroDivisionError, TypeError):
    print("¡Error! Operación no válida.")
```

### Acceder al objeto de error

Puedes usar as para acceder al mensaje del error:

```python
# Ejemplo 2.5
try:
    resultado = 10 / 0
except ZeroDivisionError as e:
    print("Detalles del error:", e)
```

### Uso de else
El bloque else se ejecuta si no ocurre ningún error:

```python
# Ejemplo 2.5

try:
    resultado = 10 / 2
except ZeroDivisionError:
    print("Error de división.")
else:
    print("El resultado es:", resultado)
```

### Uso de finally

El bloque finally se ejecuta siempre: haya error o no.
Útil para liberar recursos (archivos, conexiones, etc.).

```python
try:
    f = open("archivo.txt", "r")
    contenido = f.read()
except FileNotFoundError:
    print("El archivo no existe.")
finally:
    print("Cerrando archivo.")
```

### Uso de try/except dentro de funciones

```python
def dividir(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("¡Error! No puedes dividir entre cero.")
        return None

print("División segura:", dividir(10, 2))
print("División segura:", dividir(10, 0))
```

### Resumen
- **try**: donde va el código que puede fallar.
- **except**: para manejar errores específicos.
- **else**: se ejecuta si no hubo errores.
- **finally**: se ejecuta siempre, útil para cerrar recursos.
