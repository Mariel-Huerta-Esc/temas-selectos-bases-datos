# 1. Obtención de datos (Consumo de la API)

Selecciona una API pública de libre acceso.

Realiza la petición utilizando la librería requests y convierte la respuesta a un DataFrame de Pandas.

```python
import requests
import pandas as pd

url = "https://rickandmortyapi.com/api/character"
respuesta = requests.get(url)
datos = respuesta.json()

df = pd.DataFrame(datos["results"])

print(df["name"].head())
print(df.dtypes)
```

# 2. Diagnóstico y transformación de variables (Importante si tus datos son de texto, verifica los tipos de datos)

Utiliza .info() y .describe() para examinar tu DataFrame.

¿Qué pasa si tus datos son muy textuales? Si la API elegida devuelve principalmente texto (nombres, descripciones, categorías), deberás preparar y transformar tus datos antes de analizarlos. Cuentas con dos opciones (puedes aplicar ambas):

          - Conversión de tipos: Si hay números o fechas disfrazados de texto, conviértelos al tipo correcto (int, float, datetime) usando .astype() o pd.to_numeric().

           -Creación de variables numéricas (Ingeniería de características): Genera nuevas columnas numéricas a partir del texto. Por ejemplo: 
           
           -calcular la longitud de los caracteres de un campo de texto (como el largo de un título usando .apply(len)), contar cuántas palabras tiene una descripción, o contar la frecuencia de elementos en una lista anidada.

```python
df.info()
df.describe()
```

```python
#Como en su mayoría es texto lo que me devuelve entonces voy a hacer:  Creación de variables numéricas

#la idea es contar cuantos carácteres tiene cada elemento para poder tener datos numéricos:

df["conversion_name"] = df["name"].apply(len)


df["conversion_status"] = df["status"].apply(len)

df["conversion_species"] = df["species"].apply(len)

df["conversion_type"] = df["type"].apply(len)

df["conversion_gender"] = df["gender"].apply(len)

df["conversion_origin"] = df["origin"].apply(len) #nos devuelve el numero de claves del diccionario

df["conversion_location"] = df["location"].apply(len) #nos devuelve el número de claves del diccionario

df["conversion_image"] = df["image"].apply(len)
 
df["conversion_episode"] = df["episode"].apply(len)

df["conversion_url"] = df["url"].apply(len)

df["conversion_created"] = df["created"].apply(len)
print(df[["created", "conversion_created"]]) 

#revisé absolutamente todos los prints para verificar que el conteo estuviera correcto o al menos tuviera sentido, pero solo dejé el del final
#para fines prácticos
```

```python
#voy a convertir todos los datos a string para que me deje hacer el conteo
df["conversion_name"] = df["name"].astype(str).apply(len)

print(df[["name", "conversion_name"]])
```
